```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA 

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import mutual_info_classif

from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedShuffleSplit

from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans

from imblearn.over_sampling import SMOTE
# from imblearn.under_sampling import TomekLinks
# from imblearn.over_sampling import ADASYN
# from imblearn.over_sampling import BorderlineSMOTE
```

```python
# Loading dataset
labels = ["age", "workclass", "fnlwgt", "education", "education-num",
          "marital-status", "occupation" ,"relationship", "race", "sex",
          "capital-gain", "capital-loss", "hours-per-week", "native-country", "group"]

adult_data = pd.read_csv('adult.data', names=labels,
                         sep=',\s', na_values=["?"], engine='python')
adult_data.head(100)
```

```python
#Handling Categorical data

# adult_data.info()
# cat_df = adult_data.select_dtypes(include=['object']).copy()
# cat_df.head()

def handle_non_numerical_data(df):
    columns = df.columns.values
    for column in columns:
        text_digit_vals = {}
        def convert_to_int(val):
            return text_digit_vals[val]

        if df[column].dtype != np.int64 and df[column].dtype != np.float64:
            column_contents = df[column].values.tolist()
            unique_elements = set(column_contents)
            x = 1
            for unique in unique_elements:
                if unique not in text_digit_vals:
                    text_digit_vals[unique] = x
                    x+=1

            df[column] = list(map(convert_to_int, df[column]))

    return df

handle_non_numerical_data(adult_data)
adult_data.head()
```

```python
#Feature for Each data

adult_features = adult_data.values

#Class for Each data

adult_target = adult_data['group'].values
print ("\n----- Feature for Each data ----- \n")
print(adult_features)
print ("\n----- Class for Each data ----- \n")
print(adult_target)
```

```python
#feature selection (rank) filter-based

print ("\n----- Feature for Each data ----- \n")
print(adult_features[1:5,:])
selector = SelectKBest(mutual_info_classif, k=5).fit_transform(adult_features, adult_target) 
print ("\n----- Selection ----- \n")
print (selector)
```

```python
#Using PCA with 2 Component

pca = PCA(n_components=2) #Principal component
X_pca = pca.fit_transform(adult_features)
# X_pca

adult_target[adult_target=='<=50K'] = 1
adult_target[adult_target=='>50K'] = 2


plt.scatter(X_pca[:, 0], X_pca[:, 1], marker='o', c=adult_target)
plt.show()
print ("\n----- PCA ----- \n")
print(X_pca)
```

```python
#Using LDA

lda = LDA(n_components=2)
X_lda = lda.fit(adult_features, adult_target).transform(adult_features)

adult_target[adult_target=='<=50K'] = 1
adult_target[adult_target=='>50K'] = 2

plt.scatter(X_lda[:, 0], X_lda[:, 0], marker='o', c=adult_target)
plt.show()
print ("\n----- LDA ----- \n")
print(X_lda)
```

```python
# Stratify sampling

Train, Test = train_test_split(adult_data, test_size=0.30, stratify=adult_data['group'])
print(Train.shape)
print(Test.shape)

print ("\n----- Train ----- \n")
print(Train.head())
print ("\n----- Test ----- \n")
print(Test.head())
```

```python
label1a = sum(adult_data['group']==1)
label1b = sum(Train['group']==1)
label2a = sum(adult_data['group']==2)
label2b = sum(Train['group']==2)
print('Jumlah kelas <=50K sebenarnya', label1a)
print('Jumlah kelas <=50K stratified sampling', label1b)
print('Jumlah kelas >50K sebenarnya', label2a)
print('Jumlah kelas >50K stratified sampling', label2b)
```

```python
# Clustering Sampling

imp_mode=SimpleImputer(missing_values=np.NaN, strategy='most_frequent')
imp_mode.fit(adult_data)
clean_data = pd.DataFrame(imp_mode.transform(adult_data))
clean_data.columns = adult_data.columns
clean_data.index = adult_data.index
```

```python
n_clusters = 2
test_size=0.7

data_without_class = clean_data.drop(columns='group')
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(data_without_class)
selected_index=[]
for i in range(n_clusters):

    C_i = np.where(kmeans.labels_ == i)[0].tolist() 
    n_i = len(C_i) # number of points in cluster i
    
    
    sample_i = np.random.choice(C_i, int(test_size * n_i)) 
    selected_index.append(sample_i)
print('Index yang terpilih:\n',selected_index)

train1=adult_data.iloc[selected_index[0],:]
train2=adult_data.iloc[selected_index[1],:]

print('Jumlah data dari cluster pertama',len(train1['group']))
print('Jumlah data dari cluster kedua',len(train2['group']))
```

```python

def calc(data,label):
    X_train, X_test, Y_train, Y_test = train_test_split(data, label, test_size=.2, stratify=label, random_state=0)
    clf_DT = tree.DecisionTreeClassifier(random_state=0)
    clf_DT = clf_DT.fit(X_train, Y_train)

    hasil_testing_DT = clf_DT.predict(X_test)
    accuracy_DT = accuracy_score(Y_test, hasil_testing_DT)
    print(accuracy_DT)

    scores_DT = cross_val_score(clf_DT, data, label, cv=5)
    print(scores_DT)
    print(np.mean(scores_DT))
```

```python
def convertTo2Dimension(data):
    pca = PCA(n_components=2)
    return pd.DataFrame(pca.fit_transform(data))

def generateListColor(data):
    color = []
    for single in data:
        if single == '0':
            color.append(0.3)
        else:
            color.append(0.8)
    return color

def printCount(name,data,target):
    print (name)
    print (data.shape)
    print (np.count_nonzero(target=='1'))
    print (np.count_nonzero(target=='2'))
    calc(data,target)
    print ('\n')
```
